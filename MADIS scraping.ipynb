# MADIS script

```python
from bs4 import BeautifulSoup
import requests
import time, os
from datetime import datetime as dt
from datetime import timedelta

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
```

```python
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
def rendering(url):
    driver = webdriver.Chrome(chromedriver)
    driver.get(url)
    time.sleep(10)  # wait for the web page to load
    render = driver.page_source                              
    driver.quit()                                            
    return render                                            
```

```python
def scrape_madis(start_time, end_time):
    url = 'https://madis-data.ncep.noaa.gov/public/sfcdumpguest.shtml'
    outfile = 'madis_test_full.csv'
    madis_page = rendering(url)


    if not os.path.exists(outfile):
        mode = 'w'
    else:
        mode = 'a'
        
        with open(outfile, mode) as f:
            # write column headers for each parameter into the file for later use
            if mode == 'w':
                f.write('STAID,'
                        'OBDATE,OBTIME,PVDR_SUBPVDR,'
                        'TD,D,RH,D,ALTSE,D,SLP,D,P,D,T,D,FF,D,VISIB,D,'
                        'ELEV,D,LAT,D,LON,D,PCP1H,D,PCP24H,D,')
            
            while start_time != end_time:
                # formatting search query
                time = start_time.strftime('%Y%m%d_%H%M')
                
                #selenium interactions
                search_box = driver.find_element(By.NAME, 'time')
                search_box.send_keys(time)
                
                get_latlon = driver.find_element(By.CSS_SELECTOR, "input[type='radio'][name='dfltrsel'][value='1']").click()

                sw_lat = driver.find_element(By.NAME, 'latll')
                sw_lat.clear()
                sw_lat.send_keys('25')

                sw_lon = driver.find_element(By.NAME, 'lonll')
                sw_lon.clear()
                sw_lon.send_keys('-125')

                ne_lat = driver.find_element(By.NAME, 'latur')
                ne_lat.clear()
                ne_lat.send_keys('55')

                ne_lon = driver.find_element(By.NAME, 'lonur')
                ne_lon.clear()
                ne_lon.send_keys('-60')
                
                select_providers = driver.find_element(By.CSS_SELECTOR, "input[type='radio'][name='pvdrsel'][value='1']").click()
                driver.switch_to.window(driver.window_handles[1])
                asos = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='ASOS']").click()
                other_mtr = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='OTHER-MTR']").click()
                asos_hfm = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='ASOS-HFM']").click()
                driver.switch_to.window(driver.window_handles[0])
                
                var_sel = driver.find_element(By.CSS_SELECTOR, "input[type='radio'][name='varsel'][value='1']").click()
                driver.switch_to.window(driver.window_handles[2])
                td = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='TD']").click()
                rh = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='RH']").click()
                altse = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='ALTSE']").click()
                slp = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='SLP']").click()
                p = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='P']").click()
                t = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='T']").click()
                ff = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='FF']").click()
                elev = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='ELEV']").click()
                lat = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='LAT']").click()
                lon = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='LON']").click()
                pcp1h = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='PCP1H']").click()
                pcp24h = driver.find_element(By.CSS_SELECTOR, "input[type='checkbox'][value='PCP24H']").click()
                driver.switch_to.window(driver.window_handles[0])
                
                output_sel = driver.find_element(By.CSS_SELECTOR, "input[type='radio'][name='xml'][value='2']").click()
                
                submit_button = driver.find_element(By.CSS_SELECTOR, "input[type='button'][value='submit']").click()
                time.sleep(20)
                
                driver.switch_to.window(driver.window_handles[3])
                madis_data = driver.find_element(By.XPATH, "//pre[1]")
                
                madis_soup = BeautifulSoup(madis_data.text, 'lxml')
                csv = madis_soup.find('p').text
                results = []
                for line in csv.split('\n')[1:-1]:
                    row = line
                    results.append(row)
                
                # madis_soup = BeautifulSoup(madis_page, 'html.parser')
                # soup_container = madis_soup.find('lib-city-history-summary')
                # soup_data = soup_container.find_all('tbody', class_='ng-star-inserted')
                
                # DO STUFF
                
                
                f.write(results)
                f.write('\n')
                
                start_time += timedelta(hours=1)
```

```python
start_time = dt(year=2021, month=3, day=29, hour=1)
end_time = dt(year=2021, month=3, day=30, hour=0)

scrape_madis(start_time, end_time)
```
